<b><h1>Exploratory Data Analysis</h1></b>

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

<b>From the above graph we can see that most of the buyers are Females</b>

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

<b>From the above graph we can see that most of the buyers are from age 26-35</b>

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.set(rc={'figure.figsize':(25,10)})

sns.barplot(x = 'State', y = 'Orders', data = sales_state)
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize':(25,10)})

sns.barplot(x = 'State', y = 'Amount', data = sales_state)
```

<b>From above graph we can see that most of the orders and total sales/amount is from Uttar Pradesh, Karnataka and Maharashtra</b>

<b><h3>Marital Status</h3></b>

```python
ax = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize':(6,5)})

sns.barplot(x = 'Marital_Status', y = 'Amount', data = sales_state, hue = 'Gender')
```

<b>From the above graph we can see that most of the buyers are married(women) and they have high purchasing power</b>



```python
ax = sns.countplot(data = df, x = 'Occupation')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize':(25,10)})

sns.barplot(x = 'Occupation', y = 'Amount', data = sales_state)
```

<b>From the above graph we can see that most of the buyers are working in IT, Aviation and Healthcare Sector</b>

```python
ax = sns.countplot(data = df, x = 'Product_Category')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize':(25,10)})

sns.barplot(x = 'Product_Category', y = 'Amount', data = sales_state)
```

<b>From the above graph we can see that most of the sold products are from Food, Footwear and Electronics Category</b>

```python
sales_state = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.set(rc={'figure.figsize':(25,10)})

sns.barplot(x = 'Product_ID', y = 'Orders', data = sales_state)
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending = False).plot(kind = 'bar')
```
